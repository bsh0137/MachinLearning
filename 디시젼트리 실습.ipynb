```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

from utilities import visualize_classifier
```

```python
# 입력 데이터
input_file = 'C:\\Users\customer\\Desktop\\머신러닝 공부\\Artificial_Intelligence_with_Python_Code\\Chapter 3\\code\\data_decision_trees.txt'
data = np.loadtxt(input_file, delimiter =',')
x, y = data[:,:-1], data[:,-1]
```

```python
# 레이블을 기준으로 입력 데이터를 두 개의 클래스로 나눈다
class_0 = np.array(x[y==0])
class_1 = np.array(x[y==1])

# 입력 데이터 시각화
plt.figure()
plt.scatter(class_0[:, 0], class_0[:,1], s=75, facecolors='black', edgecolors='black', linewidth=1, marker='x')
plt.scatter(class_1[:, 0], class_1[:,1], s=75, facecolors='white', edgecolors='black', linewidth=1, marker='o')
plt.title("Input data")
```

```python
#데이터를 학습 데이터셋과 테스트 데이터셋으로 나눔
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state=5)
train_class_0 = np.array(x_train[y_train==0])
train_class_1 = np.array(x_train[y_train==1])
test_class_0 = np.array(x_test[y_test==0])
test_class_1 = np.array(x_test[y_test==1])
plt.scatter(train_class_0[:,0], train_class_0[:,1], s=75, facecolors='black', edgecolors='black', linewidth=1, marker='x')
plt.scatter(train_class_1[:,0], train_class_1[:,1], s=75, facecolors='black', edgecolors='black', linewidth=1, marker='o')
plt.scatter(test_class_0[:,0], test_class_0[:,1], s=75, facecolors='blue', edgecolors='blue', linewidth=1, marker='x')
plt.scatter(test_class_1[:,0], test_class_1[:,1], s=75, facecolors='blue', edgecolors='blue', linewidth=1, marker='o')
plt.title("그냥 해봄")
```

```python
params = {'random_state': 0, 'max_depth' : 4}
classifier = DecisionTreeClassifier(**params)
classifier.fit(x_train,y_train)
visualize_classifier(classifier, x_train, y_train)

y_test_pred = classifier.predict(x_test)
visualize_classifier(classifier, x_test, y_test)


# 분류기 성능 평가
class_names = ['Class-0', 'Class-1']
print("\n" + "#"*40)
print("\nClassifier Performance on Training Dataset\n")
print(classification_report(y_train,classifier.predict(x_train), target_names=class_names))
print("#"*40 + "\n")
print("#"*40)
print("\nClassifier_ Performance on Test Dataset\n")
print(classification_report(y_test,classifier.predict(x_test), target_names=class_names))
print("#"*40+"\n")

plt.show()
```
